```python
# Import libraries for data analysis
import pandas as pd
import numpy as np


# Libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("superstore_final_dataset (1).csv", encoding='latin1')
df.head()

#Data Understanding
df.info()
df.describe()
df.columns

df.isnull().sum()

df.drop_duplicates(inplace=True)

# Convert 'Order_Date' to datetime objects
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d/%m/%Y')
# Extract month from 'Order_Date'
df['Month'] = df['Order_Date'].dt.month

#Key Performance Indicators
total_sales = df['Sales'].sum()
average_sales = df['Sales'].mean()
total_orders = df['Order_ID'].nunique()
total_customers = df['Customer_ID'].nunique()

print("Total Revenue:", total_sales)
print("Average Sales Value:", average_sales)
print("Total Orders:", total_orders)
print("Total Customers:", total_customers)

region_sales = df.groupby('Region')['Sales'].sum().sort_values()

region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()

category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()

#Sales by Sub-Category
sub_sales = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False)

sub_sales.head(10).plot(kind='bar')
plt.title("Top 10 Sub Categories by Sales")
plt.ylabel("Sales")
plt.show()
df.columns = df.columns.str.strip()

df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Month'] = df['Order_Date'].dt.month
df['Year'] = df['Order_Date'].dt.year
```

**Dataset: Kaggle Superstore Sales Dataset**

This dataset contains retail transaction data such as order details, product category, region, sales, quantity, discount, and profit, which can be used to analyze sales trends and product performance.

The dataset contains around 9,994 records and includes product categories such as Furniture, Office Supplies, and Technology along with regional information like East, West, Central, and South.

**Load Dataset**

The dataset was loaded into a Pandas DataFrame using the read_csv() function. The head() function was used to display the first five rows of the dataset in order to understand the structure and columns present in the dataset.

**Data Understanding**

Data exploration was performed to understand the dataset structure. The info() function was used to check the data types and identify missing values in the dataset. The describe() function provided statistical summaries such as mean, minimum and maximum values for numerical columns

**Data Cleaning**

Data cleaning was performed to ensure data quality and accuracy. Missing values were checked using the isnull() function. Duplicate records were removed using the drop_duplicates() function to avoid repetition in the analysis.

**Key Performance Indicator**

Key performance indicators were calculated to measure business performance. Total revenue represents the overall sales generated by the store. Average sales value indicates the average amount spent per transaction. The number of unique orders represents total purchase transactions. The number of unique customers indicates the size of the customer base.

**Sales by Region**

Regional sales analysis was performed to identify which region contributes the highest revenue. Sales were aggregated by region and visualized using a bar chart to compare regional performance.

**Sales by Category**

Sales performance across different product categories was analyzed. The dataset was grouped by category and total sales were calculated to determine which category generates the highest revenue.

**Sales by Sub-Category**

Product level analysis was conducted by grouping sales according to sub-categories. This helps identify the best selling product types and the products that generate the highest revenue.
